```python
def count_NAs(dfna, perc):
    NaList = []
    numOfRows = len(dfna)
    for i in dfna.columns:
        numOfNAs = dfna[i].isna().sum()
        print(f'{i:<30}, Rows: {numOfRows:>5}, NAs: {numOfNAs:<25}, precentage: {round(numOfNAs/numOfRows * 100.00,2)}')
        if ((numOfNAs/numOfRows) * 100.000) >= perc:
           NaList.append(i)
    print(f'columns with NAs over {perc}%: {NaList}')
    return NaList
# Display NAs
cutoff = 40
NAs = count_NAs(df1, cutoff)
print(f"Total with NAs over {cutoff}% column number :",len(NAs))
```

# **4. Feature Engineering:**

```python
df5['review_counts'] = df5.groupby('listing_id')['review_id'].transform('count')
```

```python
df5['avg_nights'] = (df5['maximum_nights'] + df5['minimum_nights']) / 2
```

```python
df5['month'] = pd.to_datetime(df5['date']).dt.month
```

```python
df5['host_since'] = pd.to_datetime(df5['host_since'])

df5['year_month'] = df5["host_since"].dt.to_period('M')
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()


df5['host_is_superhost']= label_encoder.fit_transform(df5['host_is_superhost'])
df5['host_has_profile_pic']= label_encoder.fit_transform(df5['host_has_profile_pic'])
df5['host_identity_verified']= label_encoder.fit_transform(df5['host_identity_verified'])
df5['instant_bookable']= label_encoder.fit_transform(df5['instant_bookable'])
# the f is 0 and the t is 1
```

```python
aggregated_data = df5.groupby('host_id').agg(
    total_price=pd.NamedAgg(column='price', aggfunc='sum'),
    average_rating=pd.NamedAgg(column='review_scores_rating', aggfunc='mean'),
    earliest_host_since=pd.NamedAgg(column='host_since', aggfunc='min'),
    total_reviews=pd.NamedAgg(column='review_id', aggfunc=lambda x: x.nunique())  # Counting unique review IDs
).reset_index()
```

```python
aggregated_data = aggregated_data.sort_values(by='total_price', ascending=False).reset_index(drop=True)
aggregated_data
```
